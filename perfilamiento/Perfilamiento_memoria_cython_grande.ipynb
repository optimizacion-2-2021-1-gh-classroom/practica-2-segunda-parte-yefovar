### memory_profiler

#### Versión 1

```python
from memory_profiler import memory_usage
import SimplexC
import numpy as np
```

```python
import random

random.seed(2)

def listaAleatorios(n,m):
      lista = [0]  * n
      for i in range(n):
          lista[i] = random.randint(0, m)
      return lista
```

```python
n=9000
m=90
```

```python
c=listaAleatorios(n,1)
b=listaAleatorios(m,1)
A_=listaAleatorios(n*m,1)
A=np.resize(np.array(A_),(m,n))
```

```python
problema = SimplexC.Simplex(c,A,b,problem='Max')
t = (problema.solve)

start_mem = memory_usage(max_usage=True)
res = memory_usage(t, max_usage=True, retval=True)
print('start mem', start_mem)
print('max mem', res[0])
print('used mem', res[0]-start_mem)
print('fun output', res[1])
```
